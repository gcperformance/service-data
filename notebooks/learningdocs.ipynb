## Introduction to Git
* Take snapshots of files over time. (Git commit graphs)
* When setting up a new repo, start with standard directory. git init starts a report.
* Can retreive and discard previous version of files.
* git diff shows difference between working tree and staging area (index).

## How to be a git expert
### Commits
* A commit tracks files: paths, permissions, contents
* Git will handle directories
* Head pointer - you are here
* Commits all have different identifiers
* No renaming in Git. It just becomes a different file.
* A commit is its entire history. 
### Branches
* Name that moves along with you as you commit
* Head points at the branch you're at
### Merging
* Git will break down commit graph. Will identify common portions, portion only accessible by local and remote. Analyzes all three branches to reconcile.
* Can identify trivial and major changes that need human review. 
* Keep working on different branches and get the version of develop when we need it.
* Fast forward merge
### Fetch and pull
* local vs origin
* Fetch, local copy of commits and renames branches
* Pull, will merge local into remote branch
* Pull is fetch and a merge
### Stash changes
* Can store as many as you want and tell git to throw changes back in
### Submodules
* Single working tree could be a git repo
* Git will track head pointer in submodule repo
* Git will not change submodules automatically
### Logs 
* Show commits in chronological order
* Can do log A B ^C filters out commits reachable by C
* log --follow file/path to find specific path
* Can filter by different things
### Bisect
* Git's superpower 
* Obvious new bug, no obvious root cause
* Helps you know when it was introduced
### Revert
* You can't share a reset
* Push a revert for bad code and everyone can pull that.
### Cherry pick
* Bugfix
### Rebase
* No relationship between commits, brand new commits
* Basically never rebase
### Git Editor
* Vim default text editor of Git
* Need to see 4 windows for merging (local, base, remote, final)
* Make a git playground, everything is broken down into subcommands and subcommand arguments
* help subcommand (git help will give you info on how to use a subcommand)
* git status / diff / add / commit typical workflow
* git add -p (for checking for bugs)
* git commit -m "my commit message"
* X^2~ (^ parent, ~ number of commits back) Relative refernces
* Turn a diff nto a commit (patch)
